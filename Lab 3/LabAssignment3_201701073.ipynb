# Helping Functions

## General Nueral Network Function

```
def neuralNetwork(weight,bias,data):
  output = tf.add(tf.matmul(weight,data) ,bias)
  return output
```

## Initialisation of Weigth and Bias matrix

```
def init(layerOneNode,layerTwoNode,layerThreeNode,outputLayer,batchSize,x):
  weight1 = tf.random.normal([layerOneNode,x], mean=0, stddev=1, dtype=tf.float32,seed=1)
  weight2 = tf.random.normal([layerTwoNode,layerOneNode], mean=0, stddev=1, dtype=tf.float32,seed=1)
  weight3 = tf.random.normal([layerThreeNode,layerTwoNode], mean=0, stddev=1, dtype=tf.float32,seed=1)
  weight4 = tf.random.normal([outputLayer,layerThreeNode], mean=0, stddev=1, dtype=tf.float32,seed=1)

  bias1 = tf.random.normal([layerOneNode,1],mean=0, stddev=1,seed=1,dtype=tf.float32,)
  bias1 = tf.repeat(bias1, repeats=batchSize,axis=1)

  bias2 = tf.random.normal([layerTwoNode,1],mean=0, stddev=1,seed=1,dtype=tf.float32,)
  bias2 = tf.repeat(bias2, repeats=batchSize,axis=1)

  bias3 = tf.random.normal([layerThreeNode,1],mean=0, stddev=1,seed=1,dtype=tf.float32,)
  bias3 = tf.repeat(bias3, repeats=batchSize,axis=1)

  bias4 = tf.random.normal([outputLayer,1],mean=0, stddev=1,seed=1,dtype=tf.float32,)
  bias4 = tf.repeat(bias4, repeats=batchSize,axis=1)
  return weight1,weight2,weight3,weight4,bias1,bias2,bias3,bias4
```

## Derivative Functions

```
def derivateSigmoid(Z):
  return tf.math.multiply(tf.math.sigmoid(Z),1-tf.math.sigmoid(Z))
```

```
def derivateRelu(Z):
  relu = np.zeros((Z.shape[0],Z.shape[1]),dtype=int)
  Z = Z.numpy()
  for i in range(Z.shape[0]):
    for j in range(Z.shape[1]):
      if(Z[i][j]>0):
        relu[i][j] = 1;
      
  return tf.convert_to_tensor(relu,dtype=tf.dtypes.float32)
```

## Loading Trainnig and Testing data

```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import time as time
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data(path="mnist.npz")
data = np.array(x_train)
data = data.reshape([60000,784])
data = data.transpose()
data = tf.convert_to_tensor(data,dtype=tf.dtypes.float32)
```

```
testData = np.array(x_test)
testData = testData.reshape([10000,784])
testData = testData.transpose()
```

## Actication Functions and analysis of model 

I have used 4 layer (3 hidden + output layer) nn. In order to set the activation function, I have different choices such as sigmoid, sofmax, RELU, tanh, etc. Since Relu has no upper positive bound, it might happen that RELU gives a large value which is then passed with sigmoid function setting the value as 1. This will create a major problem as all cell of A4 matrix will have 1 value after forward nn. Thus I have used sigmoid in hidden layers and softmax in output layer. 

```
layerOneNode = 512
layerTwoNode = 512
layerThreeNode = 256
outputLayer = 10
batchSize = 300 #divisible by 60,000
totalIteration = data.shape[1]//batchSize
alpha = 0.1
epochs = 100
acc = []
#calculate weight and bias on random basis
[weight1,weight2,weight3,weight4,bias1,bias2,bias3,bias4]= init(layerOneNode,layerTwoNode,layerThreeNode,outputLayer,batchSize,data.shape[0])

for k in range(epochs):
  for i in range(1,totalIteration+1):

    #printing the epochs number and batchNumber
    #print('\r', k,i, end='')
    x = data[:,batchSize*(i-1):batchSize*i]
    y = y_train[batchSize*(i-1):batchSize*i]
    true_output = np.zeros([outputLayer, batchSize]);
    for j in range(batchSize):
      true_output[y[j]][j] = 1
    true_output = tf.convert_to_tensor(true_output, dtype=None, dtype_hint=tf.dtypes.float32, name=None)
    one = tf.ones([outputLayer, batchSize], dtype=tf.dtypes.float32)

    
    #3 hidden layer implementation with sigmoid activation function
    Z1 = neuralNetwork(weight1,bias1,x)
    A1 = tf.math.sigmoid(Z1)
    Z2 = neuralNetwork(weight2,bias2,A1)
    A2 = tf.math.sigmoid(Z2)
    Z3 = neuralNetwork(weight3,bias3,A2)
    A3 = tf.math.sigmoid(Z3)
    Z4 = neuralNetwork(weight4,bias4,A3)
    A4 = tf.nn.softmax(Z4,axis = 0)
    
    #last layer
    #dA4 = tf.math.divide(A4 - true_output,tf.math.multiply(A4,(one-A4)))
    dZ4 = tf.math.subtract(A4,true_output)
    dW4 = tf.math.divide(tf.matmul(dZ4,A3,transpose_b=True),batchSize)
    dB4 = tf.math.divide(tf.Variable(dZ4),batchSize)
    
    #last second layer
    dA3 = tf.matmul(weight4,dZ4,transpose_a=True)
    dZ3 = tf.math.multiply( dA3 ,derivateSigmoid(Z3) )
    dB3 = tf.math.divide(tf.Variable(dZ3),batchSize)
    dW3 = tf.math.divide(tf.matmul(dZ3,A2,transpose_b=True),batchSize)
    
    #last third hidden layer or second hidden layer
    dA2 = tf.matmul(weight3,dZ3,transpose_a=True)
    dZ2 = tf.math.multiply(dA2,derivateSigmoid(Z2))
    dB2 = tf.math.divide(tf.Variable(dZ2),batchSize)
    dW2 = tf.math.divide(tf.matmul(dZ2,A1,transpose_b=True),batchSize)
  
    #first hidden layer
    dA1 = tf.matmul(weight2,dZ2,transpose_a=True)
    dZ1 = tf.math.multiply(dA1,derivateSigmoid(Z1))
    dB1 = tf.math.divide(tf.Variable(dZ1),batchSize)
    dW1 = tf.math.divide(tf.matmul(dZ1,x,transpose_b=True),batchSize)

    #Changing the weights
    weight1 = tf.math.subtract(weight1,tf.math.scalar_mul(alpha,dW1))
    weight2 = tf.math.subtract(weight2,tf.math.scalar_mul(alpha,dW2))
    weight3 = tf.math.subtract(weight3,tf.math.scalar_mul(alpha,dW3))
    weight4 = tf.math.subtract(weight4,tf.math.scalar_mul(alpha,dW4))

    #changing the bias
    bias1 = tf.math.subtract(bias1, tf.math.scalar_mul(alpha,dB1))
    bias2 = tf.math.subtract(bias2, tf.math.scalar_mul(alpha,dB2))
    bias3 = tf.math.subtract(bias3, tf.math.scalar_mul(alpha,dB3))
    bias4 = tf.math.subtract(bias4, tf.math.scalar_mul(alpha,dB4))

  sum = 0 
  a = np.zeros([10,10])
  a = tf.Variable(a,dtype=tf.int32)
  for i in range(60000//batchSize):
    result = data[:,batchSize*i:batchSize*(i+1)]
    tZ1 = neuralNetwork(weight1,bias1,result)
    tA1 = tf.math.sigmoid(tZ1)
    tZ2 = neuralNetwork(weight2,bias2,tA1)
    tA2 = tf.math.sigmoid(tZ2)
    tZ3 = neuralNetwork(weight3,bias3,tA2)
    tA3 = tf.math.sigmoid(tZ3)
    tZ4 = neuralNetwork(weight4,bias4,tA3)
    tA4 = tf.nn.softmax(tZ4)
    final = tf.math.argmax(tf.transpose(tA4),1)
    a = tf.math.add(a,tf.math.confusion_matrix(y_train[batchSize*i:batchSize*(i+1)],final))

  sum = tf.linalg.trace(a)
  s = (60000//batchSize)*batchSize
  tf.print("Accuracy in epochs : "+str(k+1)+" is ", sum*100/s)
  acc = acc + [sum*100/s]

  #Visulising weigth matrix after every 10 epochs
  if(k%10 ==0):
    fig = plt.figure(figsize=(12,3))
    ax1 = fig.add_subplot(1,4,1)
    ax1.matshow(tf.reshape(weight1[0],[28,28]),cmap=plt.cm.gray)
    ax2 = fig.add_subplot(1,4,2)
    ax2.matshow(weight2,cmap=plt.cm.gray)
    ax3 = fig.add_subplot(1,4,3)
    ax3.matshow(weight3,cmap=plt.cm.gray)
    ax4 = fig.add_subplot(1,4,4)
    ax4.matshow(weight4[:,0:30],cmap=plt.cm.gray,interpolation='nearest')
    ax1.set_xticks(())
    ax1.set_yticks(())
    ax2.set_xticks(())
    ax2.set_yticks(())
    ax3.set_xticks(())
    ax3.set_yticks(())
    ax4.set_xticks(())
    ax4.set_yticks(())
    plt.tight_layout()
    plt.show()
    

  
```

Here I plotted 4 graphs at every 10 epochs. 
1. Weight1[0] => one node of the first layer => (784,1) => Reshaped to (28,28)
2. Weight2 => hidden layer 2 all nodes all weights 
3. Weight3 => hidden layer 3 all nodes all weights
4. Weight4 => output layer => (10,batchSize) => I have plotted it for the 30 inputs => (10,30) => each column should have one bright cell denoding that nn has clasified that example to index of white cell(brigth cell)

```
plt.plot(acc)
plt.title('Epochs vs Accuracy')
plt.xlabel('Epochs count')
plt.ylabel('Accuracy')
plt.show()
```

The above graph shows how model was trained. When model was started with random number, it recognise 50% of the images correctly.
```
layerOneNode = 512
layerTwoNode = 512
layerThreeNode = 256
outputLayer = 10
batchSize = 250 #divisible by 50,000
totalIteration = data2.shape[1]//batchSize
alpha = 0.3
epochs = 100
#calculate weight and bias on random basis
[weight1,weight2,weight3,weight4,bias1,bias2,bias3,bias4]= init(layerOneNode,layerTwoNode,layerThreeNode,outputLayer,batchSize,data2.shape[0])

#splitting the whole data in specific block size
#x_batches = tf.split(data2, num_or_size_splits=int(50000/batchSize), axis=1)
#y_batches = tf.split(y_train, num_or_size_splits=int(50000/batchSize), axis=0)

for k in range(epochs):
  for i in range(1,totalIteration+1):

    #printing the epochs number and batchNumber
    print('\r', k,i, end='')
    x = data2[:,batchSize*(i-1):batchSize*i]
    y = y_train2[batchSize*(i-1):batchSize*i]
    true_output = np.zeros([outputLayer, batchSize]);
    for j in range(batchSize):
      true_output[int(y[j])][j] = 1
    true_output = tf.convert_to_tensor(true_output, dtype=None, dtype_hint=tf.dtypes.float32, name=None)
    one = tf.ones([outputLayer, batchSize], dtype=tf.dtypes.float32)


    #3 hidden layer implementation with sigmoid activation function
    Z1 = neuralNetwork(weight1,bias1,x)
    A1 = tf.math.sigmoid(Z1)
    Z2 = neuralNetwork(weight2,bias2,A1)
    A2 = tf.math.sigmoid(Z2)
    Z3 = neuralNetwork(weight3,bias3,A2)
    A3 = tf.nn.sigmoid(Z3)
    Z4 = neuralNetwork(weight4,bias4,A3)
    A4 = tf.nn.softmax(Z4,axis = 0)
    
    #last layer
    #dA4 = tf.math.divide(A4 - true_output,tf.math.multiply(A4,(one-A4)))
    dZ4 = tf.math.subtract(A4,true_output)
    dW4 = tf.math.divide(tf.matmul(dZ4,A3,transpose_b=True),batchSize)
    dB4 = tf.math.divide(tf.Variable(dZ4),batchSize)
    
    #last second layer
    dA3 = tf.matmul(weight4,dZ4,transpose_a=True)
    dZ3 = tf.math.multiply( dA3 ,derivateSigmoid(Z3) )
    dB3 = tf.math.divide(tf.Variable(dZ3),batchSize)
    dW3 = tf.math.divide(tf.matmul(dZ3,A2,transpose_b=True),batchSize)
    
    #last third hidden layer or second hidden layer
    dA2 = tf.matmul(weight3,dZ3,transpose_a=True)
    dZ2 = tf.math.multiply(dA2,derivateSigmoid(Z2))
    dB2 = tf.math.divide(tf.Variable(dZ2),batchSize)
    dW2 = tf.math.divide(tf.matmul(dZ2,A1,transpose_b=True),batchSize)
  
    #first hidden layer
    dA1 = tf.matmul(weight2,dZ2,transpose_a=True)
    dZ1 = tf.math.multiply(dA1,derivateSigmoid(Z1))
    dB1 = tf.math.divide(tf.Variable(dZ1),batchSize)
    dW1 = tf.math.divide(tf.matmul(dZ1,x,transpose_b=True),batchSize)

    weight1 = tf.math.subtract(weight1,tf.math.scalar_mul(alpha,dW1))
    weight2 = tf.math.subtract(weight2,tf.math.scalar_mul(alpha,dW2))
    weight3 = tf.math.subtract(weight3,tf.math.scalar_mul(alpha,dW3))
    weight4 = tf.math.subtract(weight4,tf.math.scalar_mul(alpha,dW4))

    bias1 = tf.math.subtract(bias1, tf.math.scalar_mul(alpha,dB1))
    bias2 = tf.math.subtract(bias2, tf.math.scalar_mul(alpha,dB2))
    bias3 = tf.math.subtract(bias3, tf.math.scalar_mul(alpha,dB3))
    bias4 = tf.math.subtract(bias4, tf.math.scalar_mul(alpha,dB4))


  sum = 0 
  a = np.zeros([10,10])
  a = tf.Variable(a,dtype=tf.int32)
  for gg in range(50000//batchSize):
    result = data2[:,batchSize*gg:batchSize*(gg+1)]
    tZ1 = neuralNetwork(weight1,bias1,result)
    tA1 = tf.math.sigmoid(tZ1)
    tZ2 = neuralNetwork(weight2,bias2,tA1)
    tA2 = tf.math.sigmoid(tZ2)
    tZ3 = neuralNetwork(weight3,bias3,tA2)
    tA3 = tf.math.sigmoid(tZ3)
    tZ4 = neuralNetwork(weight4,bias4,tA3)
    tA4 = tf.nn.softmax(tZ4)
    final = tf.math.argmax(tf.transpose(tA4),1)
    a = tf.math.add(a,tf.math.confusion_matrix(y_train2[batchSize*gg:batchSize*(gg+1)],final))

  sum = tf.linalg.trace(a)
  s = (50000//batchSize)*batchSize
  tf.print(" Accuracy : ", sum*100/s)
```
